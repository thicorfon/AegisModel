```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
skeleton = pd.read_csv('./skeleton.csv',index_col = 0)
```

```python
accelerometer = pd.read_csv('/media/josuzuki/Jo/agg_datasets/Accelerometer.csv',index_col = 0)
```

```python
accelerometer['SessionID'] = np.floor(accelerometer.ActivityID/10e6).astype(int)
```

```python
skeleton = skeleton.reset_index()
```

```python
accelerometerWithWindows = []
j = 0
for _,row in accelerometer.iterrows():
    foundValue = False
    i = 0
    skeletonWithSessionId = skeleton[skeleton.SessionID == row.SessionID]
    while foundValue == False and i < len(skeletonWithSessionId):
        if (row.Systime >= skeletonWithSessionId.iloc[i].WindowStart and row.Systime < skeletonWithSessionId.iloc[i].WindowEnd):
            withWindowsRow = {}
            withWindowsRow['Systime'] = row.Systime
            withWindowsRow['EventTime'] = row.EventTime
            withWindowsRow['ActivityID'] = row.ActivityID
            withWindowsRow['X'] = row.X
            withWindowsRow['Y'] = row.Y
            withWindowsRow['Z'] = row.Z
            withWindowsRow['Phone_orientation'] = row.Phone_orientation
            withWindowsRow['SessionID'] = row.SessionID
            withWindowsRow['WindowNumber'] = skeletonWithSessionId.iloc[i].WindowNumber
            accelerometerWithWindows.append(withWindowsRow)
            foundValue = True
        i = i + 1
        if i == 10000:
            break
    j = j + 1
    if j == 10000:
        break
    
```

```python
aux = pd.DataFrame(accelerometerWithWindows)
aux.Systime = aux.Systime.astype(int)
aux
```
