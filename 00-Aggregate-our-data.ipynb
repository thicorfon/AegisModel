```python
import pandas as pd
import os
```

```python
our_data_root = './../Our Data/'
```

```python
list_of_sessions = next(os.walk(our_data_root))[1]
```

```python
list_of_sensors = ['Accelerometer',
                  'Gyroscope',
                  'Magnetometer',
                  'KeyPressEvent',
                  'KeyboardTouchEvent']
```

```python
columns = {'Accelerometer':['Systime','EventTime','ActivityID','X','Y','Z','Phone_orientation'],
           'Gyroscope':['Systime','EventTime','ActivityID','X','Y','Z','Phone_orientation'],
           'Magnetometer':['Systime','EventTime','ActivityID','X','Y','Z','Phone_orientation'],
           'KeyboardTouchEvent':['Systime','EventTime','ActivityID','Pointer_count','PointerID','ActionID','X','Y','Pressure','Contact_size','Phone_orientation'],
           'KeyPressEvent':['Systime','PressTime','PressType','ActivityID','KeyID','Phone_orientation']}
```

```python
accelerometer = pd.DataFrame(columns=columns['Accelerometer'])
for session in list_of_sessions:
    accelerometer = accelerometer.append(pd.read_csv(our_data_root+session+'/Accelerometer.csv',names=columns['Accelerometer']))
accelerometer.to_csv(our_data_root+'Accelerometer.csv')
gyroscope = pd.DataFrame(columns=columns['Gyroscope'])
for session in list_of_sessions:
    gyroscope = gyroscope.append(pd.read_csv(our_data_root+session+'/Gyroscope.csv',names=columns['Gyroscope']))
gyroscope.to_csv(our_data_root+'Gyroscope.csv')
Magnetometer = pd.DataFrame(columns=columns['Magnetometer'])
for session in list_of_sessions:
    Magnetometer = Magnetometer.append(pd.read_csv(our_data_root+session+'/Magnetometer.csv',names=columns['Magnetometer']))
Magnetometer.to_csv(our_data_root+'Magnetometer.csv')
KeyboardTouchEvent = pd.DataFrame(columns=columns['KeyboardTouchEvent'])
for session in list_of_sessions:
    KeyboardTouchEvent = KeyboardTouchEvent.append(pd.read_csv(our_data_root+session+'/KeyboardTouchEvent.csv',names=columns['KeyboardTouchEvent']))
KeyboardTouchEvent.to_csv(our_data_root+'KeyboardTouchEvent.csv')
KeyPressEvent = pd.DataFrame(columns=columns['KeyPressEvent'])
for session in list_of_sessions:
    KeyPressEvent = KeyPressEvent.append(pd.read_csv(our_data_root+session+'/KeyPressEvent.csv',names=columns['KeyPressEvent']))
KeyPressEvent.to_csv(our_data_root+'KeyPressEvent.csv')
```
