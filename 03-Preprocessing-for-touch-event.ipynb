```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
file_name = 'TouchEvent'
df = pd.read_csv('../Our Data Windowed/' + file_name + '.csv')
```

```python
df = df[df.WindowNumber != -1]
```

```python
grouped_df = df.groupby(['SessionID', 'WindowNumber']).agg({
    'Contact_size': {
        'Contact_size_mean':'mean',
        'Contact_size_std': 'std'
    },
    'Pressure': {
        'Pressure_mean':'mean',
        'Pressure_std': 'std'
    },
})
```

```python
grouped_df
```

```python
grouped_df.columns = grouped_df.columns.droplevel()
```

```python
grouped_df = grouped_df.reset_index()
```

```python
grouped_df
```

```python
grouped_df.to_csv('../Our Data Features/TouchEvent' + '_mean_std.csv', index=False)
```
