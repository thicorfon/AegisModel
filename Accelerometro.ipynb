```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
skeleton = pd.read_csv('./skeleton.csv',index_col = 0)
```

```python
accelerometer = pd.read_csv('/media/josuzuki/Jo/agg_datasets/Accelerometer.csv',index_col = 0)
```

```python
accelerometer['SessionID'] = np.floor(accelerometer.ActivityID/10e6).astype(int)
```

```python
skeleton = skeleton.reset_index()
```

```python
accelerometer
```

```python
aux_list = []
last_session = -1
last_windowNumber = -1
last_windowEnd = -1
count = 0
j = 0
for _,row in accelerometer.iterrows():
    if row.SessionID != last_session:
        skeletonWithSessionId = skeleton[skeleton.SessionID == row.SessionID]
        last_session = row.SessionID
    currentSystime = row.Systime
    if currentSystime > last_windowEnd:
        final_df = skeletonWithSessionId[(skeletonWithSessionId.WindowStart <= currentSystime) & (skeletonWithSessionId.WindowEnd > currentSystime)]
        if final_df.empty:
            last_windowEnd = -1
            last_windowNumber = -1
        else:
            last_windowEnd = final_df.WindowEnd.values[0]
            last_windowNumber = final_df.WindowNumber.values[0]
    aux_list.append(last_windowNumber)
    count = count + 1
    j = j + 1
    if (j == 100000):
        print(count)
        j = 0
accelerometer['WindowNumber'] = aux_list
```

```python
accelerometer.to_csv('accelerometer_with_windows', index=False, compression='gzip')
```
