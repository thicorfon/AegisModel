```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_name = 'magnetometer_with_windows'
df = pd.read_csv('./' + file_name + '.csv')
```

```python
df = df[df.WindowNumber != -1]
```

```python
df
```

```python
grouped_df = df.groupby(['SessionID', 'WindowNumber']).agg({
    'X': {
        'X_mean':'mean',
        'X_std': 'std'
    },
    'Y': {
        'Y_mean':'mean',
        'Y_std':'std'
    },
    'Z': {
        'Z_mean':'mean',
        'Z_std':'std'
    }
})
```

```python
grouped_df.columns = grouped_df.columns.droplevel()
```

```python
grouped_df = grouped_df.reset_index()
```

```python
grouped_df
```

```python
grouped_df.to_csv(file_name + '_mean_std.csv', index=False)
```
